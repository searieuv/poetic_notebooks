```python
import random

spaces = False
```

```python
letter_shifts = [
    ('a', 'e'),
    ('A', 'E'),
    ('e', 'i'),
    ('E', 'I'),
    ('i', 'o'),
    ('I', 'O'),
    ('o', 'u'),
    ('O', 'U'),
    ('u', 'y'),
    ('U', 'Y'),
    ('y', 'yy'),
    ('Y', 'YY')
]
```

```python
#Copyright Iron Butterfly, 1968 or so. Not mine.

phrase = '''

In a gadda da vida, honey
Don't you know that I'm lovin' you
In a gadda da vida, baby
Don't you know that I'll always be true

Oh, won't you come with me
And take my hand
Oh, won't you come with me
And walk this land
Please take my hand

In a gadda da vida, honey
Don't you know that I'm lovin' you
In a gadda da vida, baby
Don't you know that I'll always be true

Oh, won't you come with me
And take my hand
Oh, won't you come with me
And walk this land
Please take my hand

'''

phrase_orig = phrase
```

```python
phrase_alt = list(phrase)


for i, letter in enumerate(phrase_alt):
    for x in letter_shifts:
        if letter == x[0]:
            phrase_alt[i] = x[1]
            
       
```

```python
if spaces == False:

    phrase_alt_copy = phrase_alt[:]

    for letter in phrase_alt:
        if letter == ' ':
            phrase_alt_copy.remove(' ')

    phrase_alt = phrase_alt_copy
```

```python
phrase = "".join(phrase_alt)
phrase = phrase.replace('\n\n', '\n').upper()
print(phrase)
```

```python
phrase_alt = phrase.split('\n')

phrase_alt_copy = phrase_alt[:]
for el in phrase_alt:
    if el == '':
        phrase_alt_copy.remove('')
        
phrase_alt = phrase_alt_copy

phrase_alt.sort(key=len)

phrase = "\n".join(phrase_alt)
print(phrase)
```

```python
for k in range(10):
    for i in range(20):
        snippets = []
        for j in range(2):
            snippet_length = random.randrange(1, 6)
            start_position = random.randrange(len(phrase_alt_orig))
            end_position = min([start_position + snippet_length, len(phrase_alt_orig)])
            snippet = " ".join(phrase_alt_orig[start_position:end_position])
            snippets.append(snippet)
        snippets.append(random.choice(phrase_alt))

        random.shuffle(snippets)
        print(" ".join(snippets))
    print()
    
```
