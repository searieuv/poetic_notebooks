```python
import random

spaces = False
```

```python
letter_shifts = [
    ('a', 'e'),
    ('A', 'E'),
    ('e', 'i'),
    ('E', 'I'),
    ('i', 'o'),
    ('I', 'O'),
    ('o', 'u'),
    ('O', 'U'),
    ('u', 'y'),
    ('U', 'Y'),
    ('y', 'yy'),
    ('Y', 'YY')
]
```

```python
#Copyright Iron Butterfly, 1968 or so. Not mine.

phrase = '''

In a gadda da vida, honey
Don't you know that I'm lovin' you
In a gadda da vida, baby
Don't you know that I'll always be true

Oh, won't you come with me
And take my hand
Oh, won't you come with me
And walk this land
Please take my hand

In a gadda da vida, honey
Don't you know that I'm lovin' you
In a gadda da vida, baby
Don't you know that I'll always be true

Oh, won't you come with me
And take my hand
Oh, won't you come with me
And walk this land
Please take my hand

'''

phrase_orig = phrase
```

```python
phrase_alt = list(phrase)


for i, letter in enumerate(phrase_alt):
    for x in letter_shifts:
        if letter == x[0]:
            phrase_alt[i] = x[1]
            
       
```

```python
if spaces == False:

    phrase_alt_copy = phrase_alt[:]

    for letter in phrase_alt:
        if letter == ' ':
            phrase_alt_copy.remove(' ')

    phrase_alt = phrase_alt_copy
```

```python
phrase = "".join(phrase_alt)
phrase = phrase.replace('\n\n', '\n').upper()
print(phrase)
```

```python
phrase_alt = phrase.split('\n')

phrase_alt_copy = phrase_alt[:]
for el in phrase_alt:
    if el == '':
        phrase_alt_copy.remove('')
        
phrase_alt = phrase_alt_copy

phrase_alt.sort(key=len)

phrase = "\n".join(phrase_alt)
print(phrase)
```

```python
in_the_garden = '''
I come to the garden alone
While the dew is still on the roses
And the voice I hear, falling on my ear,
The Son of God discloses
Refrain: And He walks with me, and He talks with me,
And He tells me I am His own,
And the joy we share as we tarry there,
None other has ever known
He speaks, and the sound of His voice
Is so sweet the birds hush their singing
And the melody that He gave to me
Within my heart is ringing
I'd stay in the garden with Him
Tho' the night around me be falling
But He bids me go; thro' the voice of woe,
His voice to me is calling
'''

in_the_garden = in_the_garden.replace('\n', ' ')

print(in_the_garden)
```

```python
in_the_garden_list = in_the_garden.split()

print(in_the_garden_list)
```

```python
saved_poem_lines = []

all_songs = phrase_alt_orig + in_the_garden_list

for k in range(10):
    for i in range(21):
        snippets = []
        for j in range(2):
            snippet_length = random.randrange(1, 6)
            start_position = random.randrange(len(all_songs))
            end_position = min([start_position + snippet_length, len(all_songs)])
            snippet = " ".join(all_songs[start_position:end_position])
            snippets.append(snippet)
            
        snippets.append(random.choice(phrase_alt))

        random.shuffle(snippets)
        if i % 3 == 0:
            print(" ".join(snippets))
        elif i % 3 == 1:
            print("     ", " ".join(snippets), sep='')
        else:
            print("          ", " ".join(snippets), sep='')
            
        saved_poem_lines.append("".join(snippets))
        
        
    print()
```

```python
#new_small_poem = []

for j in range(20):
    for i in range(5):
        temp_line = random.choice(saved_poem_lines)
        start_position = random.randrange(len(temp_line) - 10)
        end_position = start_position + 10
        print(temp_line[start_position:end_position])
    print()
```
